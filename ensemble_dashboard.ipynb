```python
# --- LÓGICA DE IMPORTAÇÃO ---
from loterias.megasena import MegaSena
from loterias.ensemble_backtester import EnsembleBacktester
print("✅ Módulos importados com sucesso.")
```

## Configuração do Backtest
Defina os parâmetros para rodar o ensemble (MC, RF, XGB, LSTM).

```python
lottery = MegaSena()

# Parâmetros
backtester = EnsembleBacktester(
    lottery=lottery,
    range_min=1,
    range_max=60,
    draw_count=6,
    model_args={
        'n_estimators': 100,  # Para RF e XGB
        'epochs': 10          # Para LSTM (aumente para 50-100 se tiver GPU)
    }
)
```

## Execução
Rodando backtest nos últimos 10 concursos.

```python
results = backtester.run(draws_to_test=10, verbose=True)
```

## Análise de Resultados
Vamos ver como o Consenso se saiu.

```python
import pandas as pd

data = results['details']
df_res = pd.DataFrame(data)

# Extrair hits do consenso
consensus_hits = [d['hits']['con_3'] for d in data]
print(f"Média de Acertos (Consenso 3+): {sum(consensus_hits)/len(consensus_hits):.2f}")

df_res[['draw_index', 'target', 'consensus']]
```
