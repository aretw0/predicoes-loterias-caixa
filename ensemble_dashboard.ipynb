## Configura√ß√£o do Backtest
Defina os par√¢metros para rodar o ensemble (MC, RF, XGB, LSTM).

```python
lottery = MegaSena()

# Par√¢metros
backtester = EnsembleBacktester(
    lottery=lottery,
    range_min=1,
    range_max=60,
    draw_count=6,
    model_args={
        'n_estimators': 100,  # Para RF e XGB
        'epochs': 10          # Para LSTM (aumente para 50-100 se tiver GPU)
    }
)
```

## Execu√ß√£o
Rodando backtest nos √∫ltimos 10 concursos.

```python
results = backtester.run(draws_to_test=10, verbose=True)
```

## An√°lise de Resultados
Vamos ver como o Consenso se saiu.

```python
import pandas as pd

data = results['details']
df_res = pd.DataFrame(data)

# Extrair hits do consenso
consensus_hits = [d['hits']['con_3'] for d in data]
print(f"M√©dia de Acertos (Consenso 3+): {sum(consensus_hits)/len(consensus_hits):.2f}")

df_res[['draw_index', 'target', 'consensus']]
```
